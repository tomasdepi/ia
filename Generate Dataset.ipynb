```python
import pandas as pd
from pandas.io.json import json_normalize
import json

from unicodedata import *
```

```python
SPANISH = 1
ENGLISH = 2
FRENCH = 3
GERMAN = 4
JAPANESE = 5

dictFiles = {SPANISH: "spanish_words.json", ENGLISH: "english_words.json", FRENCH: "french_words.json", GERMAN: "german_words.json"}

datasetTraining = pd.DataFrame()

for language, jsonFile in dictFiles.items():
    with open(jsonFile) as file:
        data = json.load(file)
        
    
    subDataFrame = pd.DataFrame()
    subDataFrame["Word"] = data
    subDataFrame["Language"] = language
    print(len(subDataFrame))
    
    datasetTraining = pd.concat([datasetTraining, subDataFrame])
    
    
#add japanese
japaneseDataframe = pd.read_csv("japanese_words.csv", encoding='GBK')
japaneseDataframe = japaneseDataframe[["word1", "word2"]]

secondColumn = japaneseDataframe["word2"]
secondColumn = secondColumn.rename("word1")
japaneseDataframe.drop(columns=["word2"], inplace=True)
japaneseDataframe = pd.concat([japaneseDataframe, secondColumn.to_frame()])
japaneseDataframe.rename(index=str, columns={"word1": "Word"}, inplace=True)
japaneseDataframe["Language"] = JAPANESE
    
 
datasetTraining = pd.concat([datasetTraining, japaneseDataframe])

datasetTraining.to_csv("finalDataSetTraining.csv", index=False)
```

```python
def wordToList(word):
    
    lenWord = len(word)
    listOfCeros = [0] * (20-lenWord)
    l = [ord(character) for character in word]
    
    return l + listOfCeros
    
```

```python
for x in range(1,21):
    subIndex = x-1
    datasetTraining["c"+str(x)] = [wordToList(word)[subIndex] for word in datasetTraining["Word"]]
```
